###Import rr_interval into panda

```python
kangaReader = reader.Reader(gzip.open('kanga.json.gz'))
data = kangaReader.pandas(includeBurst=True)
```

##Some nice plots of heart rate

```python

hr = data.heart_rate.dropna()
figure(figsize=(15,12))
subplot(3,1,1)
hr.plot(sharex=True)
xlabel("Time"); ylabel("Heart rate (bpm)"); title("Heart rate"); ylim((40,180));

subplot(3,1,2)
hr.resample("1min").plot(sharex=True)
xlabel("Time"); ylabel("Heart rate (bpm)"); title("Heart rate (1 min average)"); ylim((40,180));

subplot(3,1,3)
hr.resample("5min").plot(sharex=True)
xlabel("Time"); ylabel("Heart rate (bpm)"); title("Heart rate (5 min average)"); ylim((40,180));
```

```python
figure(figsize=(15,21))
d = data.resample("5s")
subplot(5,1,1)
d.heart_rate.plot()
xlim(datetime(2013,12,10,22), datetime(2013,12,11,22))
ylabel("Heart rate (bpm)"); title("Heart rate");
subplot(5,1,2)
d.respiration.plot()
xlim(datetime(2013,12,10,22), datetime(2013,12,11,22))
ylabel("Respiration (bpm)"); title("Respiration");
subplot(5,1,3)
d.temperature.plot()
xlim(datetime(2013,12,10,22), datetime(2013,12,11,22))
ylabel("Temperature (C)"); title("Temperature");
subplot(5,1,4)
d.impedance.plot()
xlim(datetime(2013,12,10,22), datetime(2013,12,11,22))
ylabel("Impedance (Ohm)"); title("Impedance");ylim(0,300)
subplot(5,1,5)
rr = data.rr_interval.dropna()
sqrt(((rr - rr.shift())**2)).resample("5s").plot()
xlim(datetime(2013,12,10,22), datetime(2013,12,11,22))
ylabel("Variation (ms)"); title("Heart rate variation");
```

###Accelerometer

```python
mydata=data.resample("1min");
```

```python
mydata.describe()
```

```python
mydata["magnitude"] = sqrt(mydata.accelerometer_x**2 + mydata.accelerometer_y**2 + mydata.accelerometer_z**2);
```

```python
def upright(x):
    G=9.8
    return True if abs(degrees(arcsin(x/G))) > 30 else False
def posture(value):
    G=9.8
    print value
    x = value[mydata.accelerometer_x]
    y = value[mydata.accelerometer_y]
    z = value[mydata.accelerometer_z]
    
    angleFromLying = degrees(arcsin(x/G))
    if abs(angleFromLying) < 30:
        return "upright"
    else:
        #lying down
        angleFromSideWays = degrees(arcsin(z/G))
        if angleFromSideWays >= 45:
            return "lying on back"
        elif angleFromSideWays <= -45:
            return "lying on belly"
        else:
            #lying sideways
            if y >= 0:
                return "lying on right side"
            else:
                return "lying on left side"
mydata["upright"] = mydata.accelerometer_x.apply(upright);
#mydata["posture"] = mydata.apply(posture)
```

```python
mydata.groupby(mydata.upright).describe()
```

```python
mydata.groupby(mydata.upright).mean()
```
