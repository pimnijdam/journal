###Import rr_interval into panda
TODO: seems useful to include this in the reader, maybe include each sensor as a column in a DataFrame to really make pandas usefull with things like groupby

```python
kangaReader = reader.Reader(gzip.open('kanga.json.gz'))
data = kangaReader.pandas(filterSensors=['heart_rate', 'rr_interval', 'respiration', 'activity', 'step_count'])
#heartRate = list([(datetime.fromtimestamp(x['date']), x['value']) for x in kangaReader.rawJSON() if x['sensor_name'] == 'heart_rate'])
#rr_interval = list([(datetime.fromtimestamp(x['date']), x['value']) for x in kangaReader.rawJSON() if x['sensor_name'] == 'rr_interval'])
#accelerometer = list([(datetime.fromtimestamp(x['date']), x['value']) for x in kangaReader.rawJSON() if x['sensor_name'] == 'accelerometer (burst-mode)'])
#ecg = list([(datetime.fromtimestamp(x['date']), x['value']) for x in kangaReader.rawJSON() if x['sensor_name'] == 'ecg (burst-mode)'])
```

##Some nice plots of heart rate

```python
figure()
(60*1000.0 / data.rr_interval).resample("5min").plot()
xlabel("Time"); ylabel("Heart rate (bpm)"); title("Heart rate (5 min average)");
```

```python
figure()
(60*1000.0 / data.rr_interval).hist(range=(0,200),bins=40)
xlabel("Heart rate (bpm)"); ylabel("Frequency"); title("Heart rate histogram");xticks(range(0,200,10));
```

###Poincaré plot of rr intervals (scatter plot of subsequent intervals)
Too many points to make something useful of this graph.

```python
figure(figsize=(4,4))
scatter(data.rr_interval, data.rr_interval.shift());
xlabel("rr interval"); ylabel("subsequent rr interval"); title("Pointcare plot of rr intervals");
```

###Panda show off with heartrate variation
Shift the time series, calculate absolute difference and resample to every 5 min (how="mean")

```python
figure(figsize=(15,4))
(abs(data.rr_interval - data.rr_interval.shift())).resample("5min").plot()
ylabel("Interval variation"); title("Heart rate variation (5 min averaged)");
```
