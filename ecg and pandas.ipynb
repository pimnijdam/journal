###Poincaré plot of rr intervals (scatter plot of subsequent intervals)
Too many points to make something useful of this graph.

```python
figure(figsize=(4,4))
scatter(data.rr_interval, data.rr_interval.shift());
xlabel("rr interval"); ylabel("subsequent rr interval"); title("Pointcare plot of rr intervals");
```

###Panda show off with heartrate variation
Shift the time series, calculate absolute difference and resample to every 5 min (how="mean")

```python
figure(figsize=(15,4))
(abs(data.rr_interval - data.rr_interval.shift())).resample("5min").plot()
ylabel("Interval variation"); title("Heart rate variation (5 min averaged)");
```
