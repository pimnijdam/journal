#Some statistics for the wk2010
Use results of the pool games of wk2010 to derive some stats. These stats might be useful in predicting wk2014

```python
#poule results, in order from https://nl.wikipedia.org/wiki/Wereldkampioenschap_voetbal_2010
results = [(1, 1),
 (0, 0),
 (0, 3),
 (0, 2),
 (0, 1),
 (1, 2),
 (2, 0),
 (1, 0),
 (4, 1),
 (2, 1),
 (0, 2),
 (2, 2),
 (1, 1),
 (0, 1),
 (2, 2),
 (0, 0),
 (0, 1),
 (1, 0),
 (0, 1),
 (4, 0),
 (0, 1),
 (1, 1),
 (2, 1),
 (0, 1),
 (2, 0),
 (1, 0),
 (1, 0),
 (1, 2),
 (1, 3),
 (1, 2),
 (1, 1),
 (1, 1),
 (0, 2),
 (1, 1),
 (3, 2),
 (0, 0),
 (0, 0),
 (2, 1),
 (3, 1),
 (7, 0),
 (0, 3),
 (0, 0),
 (0, 1),
 (0, 1),
 (1, 0),
 (2, 0),
 (0, 0),
 (1, 2)]
```

```python
#Goals per category: winners, losers, draw and overall
goals = []
winner = []
loser = []
draw = []
for (home,out) in results:
    goals.append(home)
    goals.append(out)
    if home > out:
        winner.append(home)
        loser.append(out)
    elif out > home:
        winner.append(out)
        loser.append(home)
    else:
        draw.append(home)
```

```python
##Some plots on scores for the different categories
```

```python
hist(goals);xlabel('goals'); ylabel('freq'); title('Goals overall');
```

```python
hist(winner);xlabel('goals'); ylabel('freq'); title('Goals winner');
```

```python
hist(loser);xlabel('goals'); ylabel('freq'); title('Goals loser');
```

```python
hist(draw);xlabel('goals'); ylabel('freq'); title('Goals draw');
```

##Observations

- the most likely score is 0 or 1 for a team
- the most likely score for a winner is 1 or 2
- the most likely score for a loser is 0
- the most likely score for a draw is 0 or 1
