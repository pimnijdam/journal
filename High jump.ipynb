Estimate the impact of height of the person on the force to perform a jump to a certain height. Just some rough calculations to get some feeling for this.

Approach: Given a hight, calculate how high the jumper needs to move center of gravity to that height. How much energy that requires. The average force during takeoff (shorter persons have shorter legs, therefore shorter path to build up the energy, therefore higher force). Then consider impact of length of weight using constant bmi. Then consider impact of weight on strength. Eventually plot relative force to make the jump given the hight of the person.

Define some formulas to estimate force needed for a jump. Parameters:

 - mass: mass of the jumper
 - jump height
 - height of jumper
 - compression: the distance 

```python
%pylab inline
```

```python
g=9.81
def E(mass, h):
    return .5*mass*h
def F(E, dx):
    return float(E) * g / dx

def jump_force(mass, h_jumper, h_jump, c=0.15):
    j = h_jump - h_jumper/2.0
    return F(E(mass,j), h_jumper*c)
```

Plot needed force as a function of height for jumping center of gravity to 2 meter

```python
heights = arange(1.5,2.2, 0.01)
jump = 2.0
def plot_mass(mass):
    plot(heights, [jump_force(mass, x, jump) / 10 for x in heights], label="{} Kg".format(mass))
plot_mass(50)
hold(True)
plot_mass(65)
plot_mass(80)
legend()
title("Jump force to bring cog to 2 meter"); xlabel("Height of jumper (m)"); ylabel("Force (deca Newton)");
```

Taller jumpers probably also have more mass, so let's compensate for that with a constant bmi. 

```python
def mass(bmi, height):
    #bmi = m / height**2
    mass = bmi * height**2
    return mass
bmi = 20
jump = 1.9
plot(heights, [jump_force(mass(bmi,x), x, 2.0) / 10 for x in heights], label="jump 2 m")
plot(heights, [jump_force(mass(bmi,x), x, 2.2) / 10 for x in heights], label="jump 2.20 m")
legend(loc=0)
title("Jump force with constant bmi"); xlabel("Height of jumper (m)"); ylabel("Force (deca Newton)");
```

Interestingly this shows that worst length is equal to the jump height, irrespective of bmi. But also that length now is way less important, with only a small variation. As we know that small people jump less high, this is a bad model, or at least it lacks to take into account some variable.

Heavier jumpers also have more muscle. So let's plot relative force, i.e. the force to make the jump divided by the weight of the jumper. As before, we keep bmi constant.

```python
bmi = 20
plot(heights, [jump_force(mass(bmi,x), x, 2.0) / mass(bmi,x) / g for x in heights], label="jump 2 m")
plot(heights, [jump_force(mass(bmi,x), x, 2.2) / mass(bmi,x) / g for x in heights], label="jump 2.20 m")
hold(True)
legend()
title("Relative jump force "); xlabel("Height of jumper (m)"); ylabel("Relative force");
```
