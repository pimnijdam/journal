```python
%pylab inline
import pandas as pd
from datetime import datetime, timedelta

#Here credentials are stored that I don't want to share
from secrets import cs_api_key
```

###Use pandas to import data from common sense
- Use sensor id's.
- Only import the interesting columns: sensor_id, date and value.
- Strip annoying "data." from column names.
- parse dates
- index by date


```python
def importSensor(sensorId):
    #url = "http://api.sense-os.nl/sensors/data.csv?API_KEY={}&sensor_id[]=445572&sensor_id[]=445574".format(cs_api_key)
    url = "http://api.sense-os.nl/sensors/{sensorId}/data.csv?API_KEY={apiKey}".format(apiKey=cs_api_key, sensorId=sensorId)
    data = pd.read_csv(url, usecols=["data.date", "data.sensor_id", "data.value"], parse_dates=True)
    data.rename(columns=lambda name: name.replace("data.", ""), inplace=True)
    data["date"] = data.date.apply(lambda x: datetime.fromtimestamp(x))
    data.set_index(data.date, inplace=True, drop=True)
    return data
hrId = 445572
tempId = 445574
```

```python
hr = importSensor(hrId)
temp = importSensor(tempId)
temp.value.plot();
```
