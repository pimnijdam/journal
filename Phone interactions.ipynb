#Phone interactions
Let's see how often I check my phone 

```python
from reader import Reader
import pandas as pd
from datetime import datetime
from datetime import timedelta
```

```python
reader = Reader(open("data/pim_screen.json"))
data = reader.pandas()
```

```python
data['screen'] = data['screen activity_screen'].map(lambda x: 1 if x == 'on' else 0)
```

```python
#select a time period with few gaps in the data
start = datetime(2014,2,14)
end = datetime(2014,3,5)
data = data[start:end]
data.screen.plot(style='o');
```

Let's have a look at the daily pattern. These peaks are during the only three days in this dataset I was hiking. So much for enjoying nature, just a phone frenzy :-) I use the phone for navigation during the hikes, which probably explains the peaks. Well, this is an eye opener I didn't think I used my phone so much during a hike. Next time I'll try to enjoy more nature instead of worrying about navigation.

```python
dailyInteractions = data.screen.resample("1d", how=sum).fillna(value=0)
dailyInteractions.plot(style='o-')
title("Phone interaction per day")
ylabel("no phone interactions per day");
```

And of course a histogram, showing approx. 50% of the days I have between 10 and 20 interactions.

```python
hist(dailyInteractions, bins=20, range=(0,100),cumulative=True, normed=True)
title("Interactions per day cumulative histogram")
xlabel("no interactions per day"); ylabel("Relative frequency");
```

An average 21 interactions per day. Well below the 100 interactions that is claimed to be an average over the smartphone population (well, a very vague claim)

```python
dailyInteractions.mean()
```

And of course the day of the the week. Sunday the clear winner.

```python
x = data.resample("1d",how=sum).fillna(value=0)
x.groupby(x.index.dayofweek).screen.mean().plot(kind='bar');
title("Phone actions per day of the week")
xlabel("Day of week (0->monday)"); ylabel("no phone interactions per day");
```

And let's have a look at my hourly pattern. Mostly during the morning and afternoon, somewhat less in the evening and few at night.

```python
x = data.resample("1h",how=sum).fillna(value=0)
x.groupby(x.index.hour).screen.mean().plot(kind='bar');
title("Phone interactions per hour of day")
xlabel("Hour of day"); ylabel("no phone interactions per hour");
```

### Conclusion
I seem to check my iPhone approximately 21 times per day.
Caveat: when I check my phone but don't unlock it it doesn't count.
