#Phone interactions
Let's see how often I check my phone 

```python
%pylab inline
from reader import Reader
import pandas as pd
from datetime import datetime
from datetime import timedelta
import gzip
```

```python
reader = Reader(open("data/screen_activity.json"))
data = reader.pandas()
```

```python
data['screen'] = data['screen activity_screen'].map(lambda x: 1 if x == 'on' else 0)
```

```python
#inpect that there are no big gaps in the data
data.screen.plot(style='o');
```

Let's have a look at the daily pattern.

```python
dailyInteractions = data.screen.resample("1d", how=sum).fillna(value=0)
dailyInteractions.plot(style='o-')
title("Phone interaction per day")
ylabel("no phone interactions per day");
```

Let's have a look at those peaks. Seems they were mostly during non-workdays (day 5 and 6 are saturday/sunday and 2014-06-09 was Ascencion, a free day in the Netherlands). Many of these peaks seem to coincide with activities as hiking and biking, in which case I use my phone for navigation.

```python
print (dailyInteractions[dailyInteractions > 30])
print (dailyInteractions[dailyInteractions > 30].index.dayofweek)
```

And of course a histogram, showing approx. 50% of the days I have between 10 and 15 interactions with my phone.

```python
hist(dailyInteractions, bins=20, range=(0,100),cumulative=True, normed=True)
title("Interactions per day cumulative histogram")
xlabel("no interactions per day"); ylabel("Relative frequency");
```

An average 17 interactions per day. Well below the 100 interactions that is claimed to be an average over the smartphone population (well, a very vague claim)

```python
dailyInteractions.mean()
```

And of course the day of the the week. Sunday the clear winner. With a remarkable similar average during the rest of the week.

```python
x = data.resample("1d",how=sum).fillna(value=0)
x.groupby(x.index.dayofweek).screen.mean().plot(kind='bar');
title("Phone actions per day of the week")
xlabel("Day of week (0->monday)"); ylabel("no phone interactions per day");
```

And let's have a look at my hourly pattern. Never earlier than 7 o'clock (did I mention I'm not an early riser ;-)), and not many interactions after 1:00.

```python
x = data.resample("1h",how=sum).fillna(value=0)
x.groupby(x.index.hour).screen.mean().plot(kind='bar');
title("Phone interactions per hour of day")
xlabel("Hour of day"); ylabel("no phone interactions per hour");
```

### Conclusion
I seem to check my phone approximately 17 times per day.
Caveat: when I check my phone but don't unlock it it doesn't count.
