#Phone interactions
Let's see how often I check my phone 

```python
from reader import Reader
import pandas as pd
from datetime import datetime
from datetime import timedelta
```

```python
reader = Reader(open("data/pim_screen.json"))
data = reader.pandas()
                
```

```python
data['screen'] = data['screen activity_screen'].map(lambda x: True if x == 'on' else False)
```

```python
#select a time period with few gaps in the data
start = datetime(2014,2,14)
end = datetime(2014,3,5)
data = data[start:end]
data.screen.plot(style='o');
```

Let's have a look at the daily pattern. The peaks are within weekends.

```python
#Let's have a look at the daily pattern
dailyInteractions = data.screen.resample("1d", how=sum)
dailyInteractions.plot(style='o-');
```

And of course a histogram, showing approx. 50% of the days I have between 10 and 20 interactions.

```python
hist(dailyInteractions, bins=20, range=(0,100),cumulative=True, normed=True);
```

An average 21 interactions per day. Well below the 100 interactions that is claimed to be an average over the smartphone population (well, a very vague claim)

```python
dailyInteractions.mean()
```

And of course the day of the the week. Sunday the clear winner.

```python
x = data.resample("1d",how=sum)
x.groupby(x.index.dayofweek).screen.mean().plot(kind='bar');
```

__ Ai something goes wrong, as this average is at odds with the average of the dailyInteractions __

And let's have a look at my hourly pattern. Hmm, this seems to indicate that even at night (0h up to 3h) on average I check my phone hourly...

```python
x = data.resample("1h",how=sum)
x.groupby(x.index.hour).screen.mean().plot(kind='bar');
```

```python
#Ai, this seems to be double the average from dailyInteractions.mean().... TODO: investigate
x = data.resample("1h",how=sum)
sum(x.groupby(x.index.hour).screen.mean())
```

### Conclusion
I seem to check my iPhone approximately 21 times per day.
Caveat: when I check my phone but don't unlock it it doesn't count.
